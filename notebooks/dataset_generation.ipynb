```python
import pandas as pd
import random
from faker import Faker

fake = Faker("en_IN")

categories = [
    "Fashion", "Fitness", "Food", "Travel", "Gaming",
    "Education", "Finance", "Tech", "Lifestyle", "Beauty"
]

platforms = ["Instagram", "YouTube"]

cities = [
    "Mumbai", "Delhi", "Bengaluru", "Hyderabad",
    "Chennai", "Pune", "Kolkata", "Ahmedabad",
    "Jaipur", "Vijayawada"
]

age_groups = ["13-17", "18-24", "25-34", "35-44"]
genders = ["Male", "Female", "Mixed"]

data = []

for i in range(1000):

    followers = random.randint(5000, 100000)

    engagement = round(random.uniform(2, 12), 2)

    avg_likes = int(followers * (engagement / 100) * random.uniform(0.7, 1.0))

    avg_comments = random.randint(20, 500)

    posts = random.randint(4, 30)

    growth_rate = round(random.uniform(0.5, 10), 2)

    video_views = int(followers * random.uniform(0.8, 3.0))

    collab_cost = int(followers * random.uniform(0.3, 1.2))

    data.append({
        "Influencer_ID": f"INF{i+1:04}",
        "Name": fake.first_name() + str(random.randint(10,999)),
        "Platform": random.choice(platforms),
        "Followers": followers,
        "Engagement_Rate": engagement,
        "Avg_Likes": avg_likes,
        "Avg_Comments": avg_comments,
        "Posts_Per_Month": posts,
        "Category": random.choice(categories),
        "Location": random.choice(cities),
        "Collab_Cost": collab_cost,
        "Audience_Age": random.choice(age_groups),
        "Audience_Gender": random.choice(genders),
        "Growth_Rate": growth_rate,
        "Video_Views": video_views
    })

df = pd.DataFrame(data)

df.head()
```

```python
df.to_csv("../dataset/influencer_dataset.csv", index=False)

print("Dataset Saved Successfully!")
```

#### DATA PREPROCESSING

```python
df_ml = df.copy()
```

```python
from sklearn.preprocessing import LabelEncoder

le_platform = LabelEncoder()
le_category = LabelEncoder()
le_location = LabelEncoder()
le_age = LabelEncoder()
le_gender = LabelEncoder()

df_ml["Platform"] = le_platform.fit_transform(df_ml["Platform"])
df_ml["Category"] = le_category.fit_transform(df_ml["Category"])
df_ml["Location"] = le_location.fit_transform(df_ml["Location"])
df_ml["Audience_Age"] = le_age.fit_transform(df_ml["Audience_Age"])
df_ml["Audience_Gender"] = le_gender.fit_transform(df_ml["Audience_Gender"])
```

```python
df_ml = df_ml.drop(["Influencer_ID", "Name"], axis=1)
```

```python
df_ml.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_data = scaler.fit_transform(df_ml)

scaled_data.shape
```

```python
import joblib

joblib.dump(scaler, "../models/scaler.pkl")
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []

for k in range(1, 11):
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )
    
    kmeans.fit(scaled_data)
    
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1,11), inertia, marker="o")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

```python
kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=10
)

clusters = kmeans.fit_predict(scaled_data)

df_ml["Cluster"] = clusters
```

```python
df_ml["Cluster"].value_counts()
```

```python
import joblib

joblib.dump(kmeans, "../models/kmeans.pkl")
```

```python
cluster_summary = df_ml.groupby("Cluster").mean()

cluster_summary
```

Cluster 0 has lower follower counts and lower costs.

Cluster 1 has higher engagement rates.

Cluster 2 contains premium influencers.

Cluster 3 shows strong growth rates.

```python
df["Cluster"] = clusters
```

```python
df.to_csv(
    "../dataset/influencer_clustered.csv",
    index=False
)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
X = df_ml.drop("Cluster", axis=1)
```

```python
nn_model = NearestNeighbors(
    n_neighbors=5,
    metric="euclidean"
)

nn_model.fit(X)
```

```python
import joblib

joblib.dump(
    nn_model,
    "../models/nearest_neighbors.pkl"
)
```

```python
sample = X.iloc[0].values.reshape(1, -1)

distances, indices = nn_model.kneighbors(sample)

print(indices)
```

```python
recommended = df.iloc[indices[0]]

recommended
```

```python
def recommend_influencers(index):

    sample = X.iloc[index].values.reshape(1, -1)

    distances, indices = nn_model.kneighbors(sample)

    return df.iloc[indices[0]]
```

```python
recommend_influencers(25)
```
